```python
import cv2

# Created a function called count_vehicles that takes in an input file 
def count_vehicles(video_name):
    vid=cv2.VideoCapture(video_name)
    if (vid.isOpened()== False): 
          print("Error opening video file")
    initial_frame = None
    counter = 0
    count_line_postion=30
    detect=[]
    offset=6
    while True:
        check, frame = vid.read()
        # If the video still has frames, 
        # it will proceed to the convert the frames to grayscale and also use gaussian blur for the noise reduction
        if check == True:
            grayscale_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
            gaussian_blur_frame=cv2.GaussianBlur(grayscale_frame,(25,25), 0)
            # The first frame captured is the baseline image
            if initial_frame is None:
            
                initial_frame = gaussian_blur_frame
                continue
            # Frame differencing between the baseline frame and the new frame using absdiff
            # As the main focus is only on the street that heads to the city centre,
            # we will splice the frames such that it will only detect that area
            delta_frame=cv2.absdiff(initial_frame[300:414,0:600],gaussian_blur_frame[300:414,0:600])
            threshold_frame=cv2.threshold(delta_frame,15,255, cv2.THRESH_BINARY)[1]
            counterShape,h =cv2.findContours(threshold_frame,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
            cv2.line(frame,(count_line_postion,0),(count_line_postion,50),(255,127,0),3)
            for c in counterShape:
                # contourArea() method filters out any small contours
                # such that cars will only be detected
                if cv2.contourArea(c) < 1400:
                    continue
                # objects that are detected will be marked out by green rectangles
                # due to the frame splicing earlier, we have to shift the rectangles by the same values
                (x, y, w, h)=cv2.boundingRect(c)
                cv2.rectangle(frame[300:], (x, y), (x+w, y+h), (0,255,0), 1)
                cx = int((x+x+w)/2)
                cy = int((y+y+h)/2)
                detect.append((cx,cy))
                cv2.circle(frame,(cx,cy),4,(0,0,255),-1)
                for (x,y) in detect:
                    if x<144 and x>139 and y<105 and y>80:
                        counter+=1
                    cv2.line(frame,(40,0),(44,500),(0,0,),3)
                    detect.remove((x,y))
                    print('car counter ', str(counter))
            cv2.putText(frame,str(counter),(400,55),cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),5)
            cv2.imshow('video orignal',frame)
            if cv2.waitKey(1) == ord('q'):
            
                break
        else:
            break
    cv2.destroyAllWindows()
    cap.release

 




        

```
